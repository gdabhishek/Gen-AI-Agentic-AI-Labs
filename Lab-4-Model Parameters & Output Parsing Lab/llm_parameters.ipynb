# LLM Parameters Lab

This notebook demonstrates the three key LLM parameters:
1. **Temperature** - Controls randomness/creativity
2. **Top P** - Controls diversity via nucleus sampling
3. **Max Tokens** - Controls maximum output length


```python
from langchain.chat_models import init_chat_model
from langchain_core.prompts import ChatPromptTemplate

# Initialize the model
llm = init_chat_model("gpt-4o-mini", model_provider="openai")
```

## 1. Temperature Parameter

**Temperature** controls the randomness of the output:
- **Low Temperature (0.0 - 0.5)**: More deterministic, focused, and consistent responses
- **High Temperature (0.7 - 1.5)**: More creative, diverse, and unpredictable responses
- **Default**: Usually around 0.7-1.0

Let's see the difference:


```python
prompt = ChatPromptTemplate.from_messages([
    ("user", "Write a creative story about a robot in one paragraph.")
])

# Low Temperature (0.1) - More deterministic
llm_low_temp = init_chat_model("gpt-4o-mini", model_provider="openai", temperature=0.1)
chain_low = prompt | llm_low_temp
response_low = chain_low.invoke({})

print("=" * 60)
print("LOW TEMPERATURE (0.1) - More Deterministic")
print("=" * 60)
print(response_low.content)
print()
```

```python
# High Temperature (1.5) - More creative
llm_high_temp = init_chat_model("gpt-4o-mini", model_provider="openai", temperature=1.2)
chain_high = prompt | llm_high_temp
response_high = chain_high.invoke({})

print("=" * 60)
print("HIGH TEMPERATURE (1.2) - More Creative")
print("=" * 60)
print(response_high.content)
```

🌡️ TEMPERATURE GUIDE:
   0.0-0.3: Factual, consistent (technical docs, analysis)
   0.4-0.8: Natural, balanced (customer service, general content)  
   0.9-2.0: Creative, experimental (brainstorming, marketing)


## 2. Top P Parameter

**Top P** (Nucleus Sampling) controls diversity by considering only tokens with cumulative probability up to P:
- **Low Top P (0.1 - 0.5)**: More focused, considers fewer token options
- **High Top P (0.8 - 1.0)**: More diverse, considers more token options
- **Default**: Usually around 0.9-1.0

Let's compare:


```python
prompt = ChatPromptTemplate.from_messages([
    ("user", "Name 5 colors.")
])

# Low Top P (0.3) - More focused
llm_low_top_p = init_chat_model("gpt-4o-mini", model_provider="openai", top_p=0.3)
chain_low_p = prompt | llm_low_top_p
response_low_p = chain_low_p.invoke({})

print("=" * 60)
print("LOW TOP P (0.3) - More Focused")
print("=" * 60)
print(response_low_p.content)
print()
```

```python
# High Top P (0.95) - More diverse
llm_high_top_p = init_chat_model("gpt-4o-mini", model_provider="openai", top_p=0.95)
chain_high_p = prompt | llm_high_top_p
response_high_p = chain_high_p.invoke({})

print("=" * 60)
print("HIGH TOP P (0.95) - More Diverse")
print("=" * 60)
print(response_high_p.content)
```

## 3. Max Tokens Parameter

**Max Tokens** limits the maximum number of tokens in the output:
- Controls the length of the response
- Prevents very long outputs
- Useful for cost control and response length management

Let's see the difference:


```python
prompt = ChatPromptTemplate.from_messages([
    ("user", '''Explain machine learning in one para less than 300 characters.
    

    ''')
])

# Low Max Tokens (50) - Short response
llm_low_max = init_chat_model("gpt-4o-mini", model_provider="openai", max_tokens=50)
chain_low_max = prompt | llm_low_max
response_low_max = chain_low_max.invoke({})

print("=" * 60)
print("LOW MAX TOKENS (50) - Short Response")
print("=" * 60)
print(response_low_max.content)
print(f"\nLength: {len(response_low_max.content)} characters")
print()
```

```python
# High Max Tokens (300) - Longer response
llm_high_max = init_chat_model("gpt-4o-mini", model_provider="openai", max_tokens=300)
chain_high_max = prompt | llm_high_max
response_high_max = chain_high_max.invoke({})

print("=" * 60)
print("HIGH MAX TOKENS (300) - Longer Response")
print("=" * 60)
print(response_high_max.content)
print(f"\nLength: {len(response_high_max.content)} characters")
```

## Summary

**Temperature**: Controls creativity/randomness
- Use low (0.1-0.3) for factual, consistent answers
- Use high (0.7-1.5) for creative writing, brainstorming

**Top P**: Controls diversity of token selection
- Use low (0.3-0.5) for focused, predictable outputs
- Use high (0.8-1.0) for diverse, varied outputs

**Max Tokens**: Controls response length
- Use low (50-100) for short, concise answers
- Use high (200-500+) for detailed explanations

**Note**: Temperature and Top P are often used together. Typically, you adjust one or the other, not both simultaneously.
